```python
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# Grid parameters
# -----------------------------
N = 5
goal = (4, 4)
pit = (2, 2)
gamma = 0.9  # discount factor
theta = 1e-4  # convergence threshold

# -----------------------------
# Reward function
# -----------------------------
def get_reward(state):
    if state == goal:
        return 10
    elif state == pit:
        return -10
    else:
        return -1

# -----------------------------
# Fixed policy: Right if possible, else Down
# -----------------------------
def policy(state):
    r, c = state
    if state in [goal, pit]:
        return None  # terminal
    if c < N - 1:
        return (r, c + 1)  # Right
    elif r < N - 1:
        return (r + 1, c)  # Down
    else:
        return state  # stay in place

# -----------------------------
# Iterative Policy Evaluation
# -----------------------------
V = np.zeros((N, N))

while True:
    delta = 0
    for r in range(N):
        for c in range(N):
            state = (r, c)
            if state in [goal, pit]:
                continue
            next_state = policy(state)
            reward = get_reward(next_state)
            v_old = V[r, c]
            V[r, c] = reward + gamma * V[next_state]
            delta = max(delta, abs(v_old - V[r, c]))
    if delta < theta:
        break

# Define terminal states
V[goal] = 10
V[pit] = -10

# -----------------------------
# Print value function
# -----------------------------
print(f"Value Function (gamma={gamma}):")
print(np.round(V, 2))

# -----------------------------
# Arrowing visualization
# -----------------------------
arrow_map = np.full((N, N), '', dtype=object)
for row in range(N):
    for col in range(N):
        if (row, col) == goal:
            arrow_map[row, col] = 'G'
        elif (row, col) == pit:
            arrow_map[row, col] = 'P'
        elif col < N - 1:
            arrow_map[row, col] = '→'
        elif row < N - 1:
            arrow_map[row, col] = '↓'

plt.imshow(V, cmap='coolwarm', interpolation='nearest')
for r in range(N):
    for c in range(N):
        plt.text(c, r, arrow_map[r, c], ha='center', va='center',
                 color='black', fontsize=14, fontweight='bold')
plt.colorbar(label='Value')
plt.title(f"Value Function & Policy (γ={gamma})")
plt.gca().invert_yaxis()
plt.show()

# -----------------------------
# Gamma sensitivity analysis
# -----------------------------
gammas = [0.5, 0.7, 0.9, 0.99]
fig, axes = plt.subplots(1, len(gammas), figsize=(16, 4))

for i, gamma_test in enumerate(gammas):
    V_test = np.zeros((N, N))
    while True:
        delta = 0
        for r in range(N):
            for c in range(N):
                if (r, c) in [goal, pit]:
                    continue
                ns = policy((r, c))
                reward = get_reward(ns)
                v_old = V_test[r, c]
                V_test[r, c] = reward + gamma_test * V_test[ns]
                delta = max(delta, abs(v_old - V_test[r, c]))
        if delta < theta:
            break
    V_test[goal] = 10
    V_test[pit] = -10

    im = axes[i].imshow(V_test, cmap='coolwarm', interpolation='nearest')
    axes[i].set_title(f"γ={gamma_test}")
    axes[i].invert_yaxis()
    fig.colorbar(im, ax=axes[i])

plt.suptitle("Gamma Sensitivity on Value Function", fontsize=16)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# Environment Setup
# -----------------------------
N = 4
goal = (3, 3)
obstacles = [(1, 1), (2, 3)]
gamma = 0.95
step_reward = -1
goal_reward = 100
theta = 1e-4

# Actions: Up, Down, Left, Right
actions = [(-1, 0), (1, 0), (0, -1), (0, 1)]
action_symbols = ["↑", "↓", "←", "→"]

# -----------------------------
# Helpers
# -----------------------------
def is_terminal(state):
    return state == goal

def is_obstacle(state):
    return state in obstacles

def get_next_state(state, action):
    if is_terminal(state) or is_obstacle(state):
        return state
    r, c = state
    dr, dc = actions[action]
    nr, nc = r + dr, c + dc
    if 0 <= nr < N and 0 <= nc < N and not is_obstacle((nr, nc)):
        return (nr, nc)
    else:
        return state  # stay if invalid

def get_reward(state):
    if state == goal:
        return goal_reward
    elif is_obstacle(state):
        return 0  # obstacle cells won't be visited in optimal policy
    else:
        return step_reward

# -----------------------------
# Policy Iteration
# -----------------------------
states = [(r, c) for r in range(N) for c in range(N)]
V = np.zeros((N, N))
policy = np.random.choice(len(actions), size=(N, N))  # random initial policy

is_policy_stable = False
iterations = 0

while not is_policy_stable:
    iterations += 1
    # --- Policy Evaluation ---
    while True:
        delta = 0
        for state in states:
            if is_terminal(state) or is_obstacle(state):
                continue
            r, c = state
            action = policy[r, c]
            next_state = get_next_state(state, action)
            reward = get_reward(next_state)
            v_old = V[r, c]
            V[r, c] = reward + gamma * V[next_state]
            delta = max(delta, abs(v_old - V[r, c]))
        if delta < theta:
            break

    # --- Policy Improvement ---
    policy_stable = True
    for state in states:
        if is_terminal(state) or is_obstacle(state):
            continue
        r, c = state
        old_action = policy[r, c]

        # Evaluate all actions
        action_values = []
        for a in range(len(actions)):
            ns = get_next_state(state, a)
            reward = get_reward(ns)
            action_values.append(reward + gamma * V[ns])

        best_action = np.argmax(action_values)
        policy[r, c] = best_action

        if best_action != old_action:
            policy_stable = False

    if policy_stable:
        is_policy_stable = True

# -----------------------------
# Results
# -----------------------------
print(f"Converged in {iterations} policy iterations.")
print("\nOptimal Value Function:")
print(np.round(V, 2))

# Create policy map
policy_map = np.full((N, N), '', dtype=object)
for r in range(N):
    for c in range(N):
        if (r, c) == goal:
            policy_map[r, c] = 'G'
        elif is_obstacle((r, c)):
            policy_map[r, c] = 'X'
        else:
            policy_map[r, c] = action_symbols[policy[r, c]]

print("\nOptimal Policy Map:")
for row in policy_map:
    print(' '.join(row))

# -----------------------------
# Find optimal path from a start point
# -----------------------------
start = (0, 0)
path = [start]
current = start
while not is_terminal(current):
    r, c = current
    if is_obstacle(current):
        break
    act = policy[r, c]
    current = get_next_state(current, act)
    if current in path:
        break  # avoid infinite loops
    path.append(current)

print(f"\nOptimal Path from {start} to goal:")
print(path)

# -----------------------------
# Visualization
# -----------------------------
fig, ax = plt.subplots()
ax.imshow(V, cmap='coolwarm', interpolation='nearest')

for r in range(N):
    for c in range(N):
        ax.text(c, r, policy_map[r, c], ha='center', va='center',
                color='black', fontsize=14, fontweight='bold')

ax.set_title("Optimal Policy & Value Function")
plt.colorbar(ax.imshow(V, cmap='coolwarm', interpolation='nearest'), label='Value')
ax.invert_yaxis()
plt.show()
```
